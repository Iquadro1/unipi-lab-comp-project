```python
import pickle
import src.bottleneck as bottleneck

# Load the results
with open('results/dragon_complexes.pkl', 'rb') as f:
    complex_res = pickle.load(f)

sample_names = list(complex_res.keys())
complex_types = list(complex_res[sample_names[0]].keys())

# Analyze each complex type across all samples
for complex_type in complex_types:
    
    # Extract this complex type from all samples
    complex_across_samples = {
        sample_name: {complex_type: complex_res[sample_name][complex_type]}
        for sample_name in sample_names
    }
    print(f"\n{'='*80}")
    print(f"BOTTLENECK DISTANCES ACROSS SAMPLES FOR: {complex_type.upper()} COMPLEX")
    print(f"{'='*80}")
    # Create a dict with sample names as keys and the complex result as values
    complex_comparison = {
        sample_name: complex_res[sample_name][complex_type]
        for sample_name in sample_names
    }
    distances = bottleneck.compute_bottleneck_distances(complex_comparison)
    mds_results = bottleneck.compute_mds(distances)
    bottleneck.visualize_bottleneck_distances(mds_results, complex_comparison.keys())
```
